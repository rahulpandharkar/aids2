```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("detailed_data.csv")
print(df.head())
print("\nBasic Statistics:")
print(df.describe())

mean_salary = df['Salary'].mean()
print(f"\nThe Average Salary is: {mean_salary:.2f}")


plt.figure(figsize=(10, 6))
average_salary_by_year = df.groupby('year')['Salary'].mean()
average_salary_by_year.plot(kind='bar', color='skyblue')
plt.title('Average Salary by Year')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
average_salary_by_region = df.groupby('region')['Salary'].mean()
average_salary_by_region.plot(kind='bar', color='lightgreen')
plt.title('Average Salary by Region')
plt.xlabel('Region')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(x='name of company', y='Salary', data=df, palette='Set2')
plt.title('Salary Distribution by Company')
plt.xlabel('Company')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()



plt.figure(figsize=(12, 6))
sns.boxplot(x='region', y='Salary', data=df, palette='Set3')
plt.title('Salary Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


region = 'MUMBAI'
region_data = df[df['region'] == region]

plt.figure(figsize=(10, 6))
plt.scatter(region_data['year'], region_data['Salary'], color='purple', alpha=0.6)
plt.title(f'Salary vs. Year for {region}')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
salary_trend = df.groupby('year')['Salary'].mean()
plt.plot(salary_trend.index, salary_trend.values, marker='o', linestyle='-')
plt.title('Overall Salary Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.grid(True)
plt.show()


print("***********************INSIGHTS******************************")
print('1. The Salary in the Year 2014 was the highest')
print('2. The city of Mumbai experienced the Higest Salary')
print('3. Salaries from year 2014 to 2017 experienced a steep fall, however it increased a little bit in the year 2017, but again carried on the steep fall in the year 2018')


```
