```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dataset.csv")

fig,ax = plt.subplots(figsize = (15, 5))

df['Manufacturer'] = df['Manufacturer'].replace('Research in Motion (RIM)', 'RIM')
df['Manufacturer'] = df['Manufacturer'].replace('Sony Ericsson', 'Sony')

sns.barplot(data = df.groupby("Manufacturer")["Units Sold (million )"].sum(), ax = ax)
```

```python
fig,ax = plt.subplots(figsize = (15, 5))
sns.barplot(data = df.groupby("Form Factor")["Units Sold (million )"].sum(), ax = ax)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(df.groupby('Smartphone?').count()['Rank'], labels = ['FALSE', 'TRUE'])


print(df.groupby('Smartphone?').count()['Rank'])
```
