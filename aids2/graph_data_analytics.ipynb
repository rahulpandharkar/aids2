```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the modified graph with new nodes and edges
G = nx.Graph()

# Add nodes
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
G.add_nodes_from(nodes)

# Add weighted edges (modifying the weights and structure)
edges = [('A', 'B', 3), ('A', 'C', 7), ('B', 'C', 2), ('B', 'D', 4),
         ('C', 'E', 6), ('D', 'E', 1), ('D', 'F', 8), ('E', 'G', 5),
         ('F', 'G', 9), ('F', 'H', 3), ('G', 'I', 4), ('H', 'I', 2)]

G.add_weighted_edges_from(edges)

# Use Dijkstra's algorithm to find the shortest path from A to I
shortest_path = nx.shortest_path(G, source='A', target='I', weight='weight', method='dijkstra')
shortest_path_weight = nx.shortest_path_length(G, source='A', target='I', weight='weight', method='dijkstra')

# Highlight the shortest path by coloring edges
edge_colors = ['red' if edge in zip(shortest_path, shortest_path[1:]) else 'gray' for edge in G.edges()]

# Define positions and plot the graph
pos = nx.spring_layout(G)  # This layout will give a new look

plt.figure(figsize=(10, 8))

# Draw edges with different colors
nx.draw_networkx_edges(G, pos, width=2, edge_color=edge_colors)

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightblue', alpha=0.8)

# Draw node labels
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black', font_weight='bold')

# Draw edge labels (weights)
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, font_color='red')

plt.title('Modified Graph with Highlighted Shortest Path')
plt.axis('off')  # Turn off axis labels
plt.show()

# Print the shortest path and its weight
print("Shortest Path:", shortest_path)
print("Shortest Path Weight:", shortest_path_weight)
```
